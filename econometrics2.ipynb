***y = B0 + B1X + E***

y= dependent variable = inidividial medical costs billed by health insurance

x= independent quantitative variable: BMI


B0: in general context it is when X is 0 it resolve the quantity of Y. So, when BMI is 0, the average of charge of medical costs is zero.


B1: The average increase of the individual charge of medical costs billed vy health insurance  when BMI rises by one unit.

```python
import statsmodels.formula.api as snf

model1=snf.ols(formula='charges ~ bmi',data=insurance).fit()
```

```python
print(model1.summary())
```
