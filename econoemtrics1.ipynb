```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation

# Generovanie údajov pre exponenciálny rast
blood_pressure = np.linspace(100, 180, 100)  # Hodnoty krvného tlaku
cv_complications = np.exp((blood_pressure - 100) / 30)  # Exponenciálny rast

# Nastavenie figúry a osi pre animáciu
fig, ax = plt.subplots(figsize=(8, 6))
ax.set_xlim(100, 180)
ax.set_ylim(0, np.max(cv_complications) * 1.1)
ax.set_xlabel("Krvný tlak (mm Hg)")
ax.set_ylabel("Kardiovaskulárne komplikácie")
ax.grid(True)

# Inicializácia línie
line, = ax.plot([], [], linewidth=2, label="Riziko kardiovaskulárnych komplikácií")

# Funkcia na inicializáciu animácie
def init():
    line.set_data([], [])
    return line,

# Funkcia na aktualizáciu animácie
def update(frame):
    line.set_data(blood_pressure[:frame], cv_complications[:frame])
    return line,

# Vytvorenie animácie
ani = animation.FuncAnimation(fig, update, frames=len(blood_pressure), init_func=init, blit=True, interval=50)

# Uloženie animácie ako video
ani.save("continuous_risk_increase.mp4", writer="ffmpeg", fps=30)

print("Animácia bola uložená ako 'continuous_risk_increase.mp4'")
```
