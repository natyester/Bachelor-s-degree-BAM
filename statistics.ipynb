```python
import matplotlib.pyplot as plt
```

```python
#Data from source: DATACOMEX
years = ['2020', '2021', '2022', '2023', '2024']
x = range(len(years))

exports_euros = [
    263628262159.03,
    314858545734.85,
    387599020129.29,
    384098393756.50,
    384464857203.31
]

imports_euros = [
    276925148710.07,
    346283414295.44,
    459202617644.58,
    423723442312.18,
    424740724715.73
]
```

```python
# Graph
width = 0.35
plt.figure(figsize=(12, 6))
plt.bar([i - width/2 for i in x], exports_euros, width=width, label='Exports (€)', color='lightblue')
plt.bar([i + width/2 for i in x], imports_euros, width=width, label='Imports (€)', color='lightgreen')
plt.plot([i - width/2 for i in x], exports_euros, color='blue', marker='o', linestyle='-', linewidth=2)
plt.plot([i + width/2 for i in x], imports_euros, color='green', marker='o', linestyle='-', linewidth=2)
plt.xlabel('Year')
plt.ylabel('Euros (€)')
plt.title(' Exports and Imports (2020–2024)')
plt.xticks(x, years)
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.legend()
plt.figtext(0.99, 0.01, 'Source: DataComex', horizontalalignment='right', fontsize=9, style='italic', color='gray')
plt.tight_layout()
plt.show()
```
